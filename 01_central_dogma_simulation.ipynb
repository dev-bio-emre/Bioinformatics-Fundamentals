### DNA Analizi ve GC KararlÄ±lÄ±ÄŸÄ±
AÅŸaÄŸÄ±daki gÃ¶rselde gÃ¶rÃ¼ldÃ¼ÄŸÃ¼ gibi, Guanin (G) ve Sitozin (C) arasÄ±nda 3 adet hidrojen baÄŸÄ± bulunur (A-T arasÄ±nda 2 adettir). Bu yÃ¼zden GC oranÄ± yÃ¼ksek DNA'lar Ä±sÄ±ya daha dayanÄ±klÄ±dÄ±r.

<div align="center">
  <h3>DNA Analizi ve GC KararlÄ±lÄ±ÄŸÄ±</h3>
  <p>Guanin (G) ve Sitozin (C) arasÄ±nda 3 adet hidrojen baÄŸÄ± bulunur. GC oranÄ± yÃ¼ksek DNA, Ä±sÄ±ya daha dayanÄ±klÄ±dÄ±r.</p>
  <img src="images\dna-base-pairing-3153.png" width="30%" />
</div>

```python
# Biopython kÃ¼tÃ¼phanesinden 'Seq' (Dizi) modÃ¼lÃ¼nÃ¼ Ã§aÄŸÄ±rÄ±yoruz

from Bio.Seq import Seq
```

```python
# 1. VERÄ°: Rastgele bir gen dizisi tanÄ±mlayalÄ±m (String olarak)
# Bu diziye biyolojik bir kimlik kazandÄ±rmak iÃ§in 'Seq' objesine Ã§eviriyoruz.

gen_dizisi = Seq("ATGCGTACGTAGCTAGCTAGCTAGCTAGCTAGC")
```

```python
# 2. ANALÄ°Z (MANUEL YÃ–NTEM): MÃ¼hendislik mantÄ±ÄŸÄ±nÄ± kuralÄ±m
# Dizinin iÃ§indeki 'G' ve 'C' harflerini sayÄ±yoruz.

g_sayisi = gen_dizisi.count("G")
c_sayisi = gen_dizisi.count("C")
```

```python
# Toplam uzunluÄŸu alÄ±yoruz

toplam_uzunluk = len(gen_dizisi)
```

```python
# YÃ¼zdeyi hesaplÄ±yoruz: (G + C) / Toplam * 100

gc_yuzdesi = (g_sayisi + c_sayisi) / toplam_uzunluk * 100
```

```python
# 3. SONUÃ‡LARI RAPORLA

print("--- Ä°SVÄ°Ã‡RE LABORATUVAR RAPORU ---")
print(f"Analiz Edilen Dizi: {gen_dizisi}")
print(f"Toplam Uzunluk: {toplam_uzunluk} nÃ¼kleotid")
print(f"G SayÄ±sÄ±: {g_sayisi}")
print(f"C SayÄ±sÄ±: {c_sayisi}")
print(f"GC OranÄ± (KararlÄ±lÄ±k): %{gc_yuzdesi:.2f}")
```

```python
# 4. YORUM (Basit bir karar mekanizmasÄ± ekleyelim)
if gc_yuzdesi > 50:
    print("SONUÃ‡: YÃ¼ksek kararlÄ±lÄ±k. Erime sÄ±caklÄ±ÄŸÄ± yÃ¼ksek.")
else:
    print("SONUÃ‡: DÃ¼ÅŸÃ¼k kararlÄ±lÄ±k. Erime sÄ±caklÄ±ÄŸÄ± dÃ¼ÅŸÃ¼k.")
```

##  AÅAMA 2: TRANSKRÄ°PSÄ°YON (DNA -> mRNA) 

```python
print("\n--- FAZ 2: mRNA DÃ–NÃœÅÃœMÃœ BAÅLIYOR ---")

# Biopython'un kendi sihirli fonksiyonunu kullanÄ±yoruz: .transcribe()
# Bu fonksiyon otomatik olarak T harflerini U yapar ve RNA objesi oluÅŸturur.

mrna_dizisi = gen_dizisi.transcribe()

print(f"Orijinal DNA: {gen_dizisi}")
print(f"Ãœretilen mRNA: {mrna_dizisi}")

# Kontrol edelim: GerÃ§ekten RNA oldu mu?

if "U" in mrna_dizisi:
    print("BAÅARILI: mRNA sentezlendi. AÅŸÄ± Ã¼retimine hazÄ±r! ğŸ’‰")
else:
    print("HATA: DÃ¶nÃ¼ÅŸÃ¼m baÅŸarÄ±sÄ±z.")
```

## AÅAMA 3: TRANSLASYON (mRNA -> PROTEÄ°N)

<div align="center">
  <h3>Genetik Kod Tablosu (Codon Table)</h3>
  <p>mRNA Ã¼zerindeki 3'lÃ¼ kodlarÄ±n Amino Asit karÅŸÄ±lÄ±klarÄ±. <b>AUG</b> baÅŸlatÄ±r, <b>STOP</b> bitirir.</p>

```python
print("\n--- FAZ 3: PROTEÄ°N SENTEZÄ° (Ä°LAÃ‡ ÃœRETÄ°MÄ°) ---")

# mRNA'yÄ± okuyup amino asit zincirine Ã§eviriyoruz.
# Ä°sviÃ§re'deki Roche veya Novartis fabrikalarÄ± bunu dev biyoreaktÃ¶rlerde yapar, biz Python'da yapÄ±yoruz.

protein_dizisi = mrna_dizisi.translate()

print(f"OluÅŸan Protein Dizisi: {protein_dizisi}")

# Biyolojik Kontrol: Protein '*' (Stop Kodon) ile bitmeli veya iÃ§ermelidir.
# '*' iÅŸareti, ribozoma "Ãœretimi durdur, protein bitti" diyen sinyaldir.

if "*" in protein_dizisi:
    print("BAÅARILI: Protein sentezlendi ve sonlandÄ±. Ä°laÃ§ Ã¼retimine hazÄ±r! ğŸ’Š")
    print("Bu dizi artÄ±k fonksiyonel bir protein olabilir, laboratuvar testlerine geÃ§ebiliriz.")
else:
    print("UYARI: Protein sentezlendi ama sonlanmadÄ±. Laboratuvar testleri gereklidir.")
```
